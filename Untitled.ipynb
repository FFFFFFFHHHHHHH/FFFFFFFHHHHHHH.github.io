## 6点之后是盗窃罪高发的时间段 

```python
tmp2 = tmp.groupby(['Category','PdDistrict']).size()
tmp2=tmp2.unstack()
tmp2.T.plot(kind="bar",figsize=(12,6))
```

```python
tmp3 = tmp.groupby(['Category','DayOfWeek']).size()
tmp3 = tmp3.unstack()
tmp3 = tmp3[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']] #本来是按照字典序进行排序，现在按照星期几进行排序
print(tmp3)
tmp3.T.plot(figsize=(10,10))
```

## 上述图得到的线趋势并不明显，因此改为每种案件在一星期内每一天的比重

```python
#每种案件在每星期种所占的比重
for i in range(6):
    tmp3.iloc[i] = tmp3.iloc[i]/tmp3.sum(axis=1)[i]
```

```python
tmp3.T.plot(figsize=(10,10))
```

## 如图，我们能够更清晰的看到犯罪类别和dayofweek之间的关系

## 接下来是月份和案件之间的关系

```python
mon_g = tmp.groupby(['Category','month']).size()
mon_g = mon_g.unstack()
for i in range(6):
    mon_g.iloc[i] = mon_g.iloc[i]/mon_g.sum(axis=1)[i]
mon_g.T.plot(figsize=(12,6),style='o-')
plt.show()
```

## 年份和案件

```python
year_g = tmp.groupby(['Category','year']).size()
year_g = year_g.unstack()
for i in range(6):
    year_g.iloc[i] = year_g.iloc[i]/year_g.sum(axis=1)[i]
plt.style.use('ggplot')
year_g.T.plot(figsize=(12,6),style='o-')
plt.show()
```

## 可见，不同种类的案件随时间是有不同变化的，如vehicle theft在05年后急剧下降，可能有专项整治等活动，theft却在12年后有升高的趋势。但是都在2015年后下降

## 数据预处理

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
#from sklearn.feature_selection import SelectKBest
#from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split
```

## 对分类目标进行类别编码 

```python
label = preprocessing.LabelEncoder()
target = label.fit_transform(train.Category)
target
```

## 去除无用和重复特征 

```python
Id = test['Id']
des = train['Descript']
res = train['Resolution']
train.drop(['Category','Descript','Resolution','Address','Dates','date'],axis=1,inplace=True)
test.drop(['Id','Address','Dates','date'],axis=1,inplace=True)
```

```python
train
```

```python
test
```

## 将DayOfweek和year,month,hour还有PdDistrict进行独热编码

```python
week_train = pd.get_dummies(train.DayOfWeek)
week_test = pd.get_dummies(test.DayOfWeek)
year_dum_train=pd.get_dummies(train.year)
month_dum_train=pd.get_dummies(train.month)
hour_dum_train=pd.get_dummies(train.hour)
year_dum_test=pd.get_dummies(test.year)
month_dum_test=pd.get_dummies(test.month)
hour_dum_test=pd.get_dummies(test.hour)
dist_train=pd.get_dummies(train.PdDistrict)
dist_test=pd.get_dummies(test.PdDistrict)
```

```python
dist_train
```

## 添加两个新特征newy,dark（根据之前的数据分析，得知年初和18点以后的犯罪率总体来说较高）

```python
train['newy'] = train['month'].apply(lambda x:1 if x==1 or x==2 else 0)
train['dark'] = train['hour'].apply(lambda x:1 if x>=18 and x<=24 else 0)
test['newy'] = test['month'].apply(lambda x:1 if x==1 or x==2 else 0)
test['dark'] = test['hour'].apply(lambda x:1 if x>=18 and x<=24 else 0)
```

## 删除特征  'month'、'hour'、'day'、'year'、'DayOfWeek'、'PdDistrict'

```python
train.drop(['month','hour','day','year','DayOfWeek','PdDistrict'],axis=1,inplace=True)
test.drop(['month','hour','day','year','DayOfWeek','PdDistrict'],axis=1,inplace=True)
```

## 将之前经过独热编码的新特征加入进去 

```python
train = pd.concat([train,week_train,dist_train,year_dum_train,month_dum_train,hour_dum_train],axis=1)
test = pd.concat([test,week_test,dist_test,year_dum_test,month_dum_test,hour_dum_test],axis=1)
```
